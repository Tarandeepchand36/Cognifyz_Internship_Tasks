```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Use absolute path to load the CSV file
file_path = r"C:\Users\tarun\Desktop\Python\Dataset .csv"
# Load the dataset 
df = pd.read_csv("Dataset .csv")
```

<p style="font-size:20px;">
Calculate the percentage of restaurants in each price range category.

```python
# Count occurrences of each price range
price_counts = df['Price range'].value_counts()
```

```python
# Calculate percentage of restaurants in each price range category
total_restaurants = len(df)
percentage_per_price = (price_counts / total_restaurants) * 100
print(percentage_per_price)
```

<p style="font-size:20px;">
Create a histogram or bar chart to visualize the distribution of price ranges among the restaurants.

```python
# Plotting the histogram or bar chart
plt.figure(figsize=(8, 6))
percentage_per_price.plot(kind='bar', color='skyblue')
plt.title('Distribution of Price Ranges Among Restaurants')
plt.xlabel('Price Range')
plt.ylabel('Percentage of Restaurants (%)')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ylim(0, 100)
```
