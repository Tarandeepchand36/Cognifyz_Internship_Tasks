```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr
```

```python
# Use absolute path to load the CSV file
file_path = r"C:\Users\tarun\Desktop\Python\Dataset .csv"
# Load the dataset 
df = pd.read_csv("Dataset .csv")
```

<p style="font-size:20px;">
Identify the restaurants with the highest and lowest number of votes.

```python
# Print the restaurant with the hightest votes

highest_votes = df.loc[df['Votes'] == df['Votes'].max()][['Restaurant Name', 'Votes']]
print(f"Restaurant with the highest votes: {highest_votes.iloc[0, 0]} with {highest_votes.iloc[0, 1]} votes")
```

```python
# Print the restaurant with the lowest votes

lowest_votes = df.loc[df['Votes'] == df['Votes'].min()][['Restaurant Name', 'Votes']]
print(f"Restaurant with the highest votes: {lowest_votes.iloc[0, 0]} with {lowest_votes.iloc[0, 1]} votes")
```

<p style="font-size:20px;">
Analyze if there is a correlation between the number of votes and the rating of a restaurant.

```python
# Assuming your DataFrame is named df and has columns 'Votes' and 'Rating'
votes = df['Votes']
ratings = df['Aggregate rating']

# Calculate the Pearson correlation coefficient
correlation, _ = pearsonr(votes, ratings)

print(f'Pearson correlation coefficient between votes and ratings: {correlation}')
```

```python
import matplotlib.pyplot as plt

plt.scatter(votes, ratings)
plt.xlabel('Votes')
plt.ylabel('Aggregate rating')
plt.title('Scatter plot of Votes vs. Rating')
plt.show()
```
