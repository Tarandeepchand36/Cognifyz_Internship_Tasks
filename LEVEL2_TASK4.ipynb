```python
import pandas as pd
import matplotlib.pyplot as plt

# Use absolute path to load the CSV file
file_path = r"C:\Users\tarun\Desktop\Python\Dataset .csv"

# Load the dataset 
df = pd.read_csv("Dataset .csv")
```

<p style="font-size:20px;">
Identify if there are any restaurant chains present in the dataset.

```python
# Identify restaurant chains by counting occurrences
chain_counts = df['Restaurant Name'].value_counts()
```

```python
# Filter out only those with multiple locations (chains)
restaurant_chains = chain_counts[chain_counts > 1].index.tolist()
```

```python
# Create a new DataFrame to aggregate ratings by chains
chain_ratings = pd.DataFrame(columns=['Chain', 'Average Rating', 'Total Outlets'])
```

```python
# Create an empty list to store chain ratings
chain_ratings = []
```

```python
# Calculate average ratings and total outlets for each chain
for chain in restaurant_chains:
    chain_data = df[df['Restaurant Name'] == chain]
    average_rating = chain_data['Aggregate rating'].mean()
    total_outlets = len(chain_data)
    chain_ratings.append({
        'Chain': chain,
        'Average Rating': average_rating,
        'Total Outlets': total_outlets
         
    })
```

```python
# Convert the list of dictionaries to a DataFrame
chain_ratings_df = pd.DataFrame(chain_ratings)
```

```python
# Sort chains by average rating
chain_ratings_sorted = chain_ratings_df.sort_values(by='Average Rating', ascending=False)
```

```python
print(chain_ratings_sorted[0:10])
```

<p style="font-size:20px;">
Analyze the ratings and popularity of different restaurant chains.

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.bar(chain_ratings_sorted['Chain'][:10], chain_ratings_sorted['Average Rating'][:10], color='skyblue')
plt.xlabel('Restaurant Chains')
plt.ylabel('Average Rating')
plt.title('Top 10 Restaurant Chains by Average Rating (Sorted)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Optionally, you can also print the sorted chains
print(chain_ratings_sorted.head())
```
