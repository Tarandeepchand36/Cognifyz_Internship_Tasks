```python
import pandas as pd
import matplotlib.pyplot as plt

# Use absolute path to load the CSV file
file_path = r"C:\Users\tarun\Desktop\Python\Dataset .csv"

# Load the dataset 
df = pd.read_csv("Dataset .csv")
```

<p style="font-size:20px;">
Analyze the distribution of aggregate ratings and determine the most common rating range.

```python
# Filter out ratings of 0
fil_Aggregate_ratings = df[df['Aggregate rating'] != 0]['Aggregate rating']
```

```python
# Analyze the distribution of aggregate ratings
plt.figure(figsize=(8, 6))
fil_Aggregate_ratings.plot(kind='hist', bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Aggregate Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

<p style="font-size:20px;">
Calculate the average number of votes received by restaurants.

```python
# Determine the most common rating range
rating_counts = fil_Aggregate_ratings.value_counts()
most_common_rating = rating_counts.idxmax()
```

```python
print(f"The most common rating range is around {most_common_rating:.2f} stars.")
```

```python
# Calculate average number of votes received by restaurants
average_votes = df['Votes'].mean()

print(f"Average number of votes received by restaurants: {average_votes:.2f}")
```
